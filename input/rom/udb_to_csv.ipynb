```python
import numpy as np
import matplotlib.pyplot as plt
import sqlite3 as lite
import csv
```

```python
db_path = '../../db/1yr.sqlite'
con = lite.connect(db_path)
con.row_factory = lite.Row
cur = con.cursor()
```

```python
# find all isotopes in db
iso = cur.execute('SELECT distinct(isotope) FROM discharge').fetchall()
iso_list = ['burnup', 'enrichment']
for i in iso:
    iso_list.append(i['isotope'])
```

```python
print(iso_list)
```

```python
assembly_id = cur.execute('SELECT distinct(assembly_id) FROM discharge GROUP BY evaluation_date, reactor_id').fetchall()
```

```python
total = len(assembly_id)
print(total)
count = 0
csv_dict = {}
with open('udb_bwr.csv', 'w') as fb:
    wrb = csv.writer(fb)
    wrb.writerow(iso_list)
    with open('udb_pwr.csv', 'w') as f:
        wr = csv.writer(f)
        wr.writerow(iso_list)
        for assembly in assembly_id:
            count += 1
            temp = np.zeros(len(iso_list), dtype=None)
            assem_id = assembly['assembly_id']
            comp = cur.execute('SELECT isotope, total_mass_g FROM discharge WHERE assembly_id = %i' %assem_id).fetchall()
            total_mass = 0
            for iso in comp:
                total_mass += float(iso['total_mass_g'])
            for iso in comp:
                indx = iso_list.index(iso['isotope'])
                # normalize it
                temp[indx] = float(iso['total_mass_g']) / total_mass
            enr_bu = cur.execute('SELECT initial_enrichment, discharge_burnup, reactor_type FROM discharge WHERE assembly_id = %i' %assem_id).fetchone()
            temp[0] = enr_bu['discharge_burnup']
            temp[1] = enr_bu['initial_enrichment']
            if enr_bu['reactor_type'] == 'BWR':
                wrb.writerow(temp)
            elif enr_bu['reactor_type'] == 'PWR':
                wr.writerow(temp)
            else:
                print('WHAT IS THIS TYPE: %s \n\n\n' %enr_bu['reactor_type'])
            print('%f %% DONE' %(float(count)/float(total) * 100))
```
