```python
# blanket and driver compositions
import numpy as np
with open('recipe_text', 'r') as f:
    lines = f.readlines()
    array = np.array(['isotope','driver_fresh', 'driver_used', 'blanket_fresh', 'blanket_used'])
    for line in lines[1:]:
        organized = ' '.join(line.split())
        organized = organized.split(' ')
        array = np.vstack((array, organized))
    isotope_array = array
```

```python
print(array[0][1])
```

```python
# build recipe file
output_file = 'recipe.xml'

output = open(output_file, 'w')

output.write('<root>\n')
for i in range(1, len(array[0])):
    output.write('<recipe>\n'
                 '\t<name>%s</name>\n'
                 '\t<basis>mass</basis>\n' %array[0][i])
    for j in range(1,len(array)):
        if float(array[j][i]) < 1e-7:
            continue
        output.write('\t<nuclide>\n'
                     '\t\t<id>%s</id>\n'
                     '\t\t<comp>%s</comp>\n'
                     '\t</nuclide>\n' %(array[j][0], array[j][i]))
    output.write('</recipe>\n\n')

output.write('</root>')

```
